```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path_retail = r"C:\Users\conta\Desktop\TASK 2 FILES ALL\RetailData_task_2_CSV.csv"
data = pd.read_csv(file_path_retail)

# 1. Explore sales trends over time
# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['MonthYear'] = data['Date'].dt.to_period('M')

# Aggregate sales by month
time_series = data.groupby('MonthYear')['Amount'].sum().reset_index()
time_series['MonthYear'] = time_series['MonthYear'].astype(str)

# Plot sales trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=time_series, x='MonthYear', y='Amount', marker='o', color='blue')
plt.title('Sales Trends Over Time', fontsize=16)
plt.xlabel('Month-Year', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Identify top-selling products and best-performing store locations
# Top-selling products
top_products = data.groupby('Itemname')['Quantity'].sum().sort_values(ascending=False).head(10)
print("Top-Selling Products:\n", top_products)

# Best-performing store locations
best_stores = data.groupby('StoreLocation')['Amount'].sum().sort_values(ascending=False).head(10)
print("Best-Performing Store Locations:\n", best_stores)

# Plot top-selling products
plt.figure(figsize=(12, 6))
top_products.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top-Selling Products', fontsize=16)
plt.xlabel('Product', fontsize=14)
plt.ylabel('Total Quantity Sold', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot best-performing store locations
plt.figure(figsize=(12, 6))
best_stores.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Best-Performing Store Locations', fontsize=16)
plt.xlabel('Store Location', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Calculate and compare average transaction amounts by location
average_transaction = data.groupby('StoreLocation')['Amount'].mean().sort_values(ascending=False)
print("Average Transaction Amount by Store Location:\n", average_transaction)

# Plot average transaction amounts
plt.figure(figsize=(12, 6))
average_transaction.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Average Transaction Amount by Store Location', fontsize=16)
plt.xlabel('Store Location', fontsize=14)
plt.ylabel('Average Transaction Amount', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Provide recommendations based on trends
print("\nRecommendations:")
print("1. Focus on promoting the top-selling products across all stores to maximize revenue.")
print("2. Increase stock and visibility of popular items in best-performing store locations.")
print("3. Analyze pricing strategies in locations with low average transaction amounts to boost sales.")
print("4. Organize promotional campaigns during months with historically low sales.")
```
